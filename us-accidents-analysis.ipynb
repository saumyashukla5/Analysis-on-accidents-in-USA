### Start Time

```python
df.Start_Time
```

### At what time of the day does accidents most frequently occur?

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

#### During Weekdays

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True
             )
```

- A high percentage of accidents occur between 6am to 10 am (probably people are in a hurry to get to work)
- Next highest percentage is 6pm to 10pm


```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

#### On Weekends

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6 ]
sns.distplot(sundays=_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

On Sundays, peak hour is between 10 am and 3pm


```python
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0 ]
sns.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

#### Which month do accidents occur the most?

```python
df_2017 = df[df.Start_Time.dt.year == 2017]
sns.distplot(df.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

We observe a spike in the number of accidents during the winter season probably due to fog and lower visibility during winters.
